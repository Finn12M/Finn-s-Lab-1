```python
import numpy as np
import matplotlib.pyplot as plt


file="satellite-failures.txt"
data = np.genfromtxt(file,skip_header=1,usecols=(1,2))
#print(data)

# how big is the array?  (rows,cols) will be returned
#print (data.shape)

#I want to select every row (:), but only those
#whose second index is 0. This gives us the first column
sunspots = data[:,0]

#I want to select every row, but only the second thing in each
#row, giving us our satellites data
satellites = data[:,1]

#print (sunspots)
#print (satellites)

intercept, slope = np.polynomial.polynomial.polyfit(sunspots,satellites,1)

print(intercept,slope)

#create best fit line's x and y points
x = np.linspace(0,160)

#best fit line
y = slope*x + intercept

plt.xlabel("# of Sunspots in a year")
plt.ylabel("# of Satellites re-entering in a year")
plt.title("Satellites burning up vs sunspot count")
plt.scatter(sunspots,satellites,color="purple")
plt.plot(x,y)

#From the plot we know the slope of the best fit line
#We can use this to estimate the number of satellites that would burn up if there was 400 sunspots in a year

i = 400
num_satellites = (0.1635072456826605*(i) + 13.95200407726704)
print("")
print("The estimated number of satellites destroyed by 400 sunspots is ", num_satellites )
print("")






```

```python
new_file = "co2.txt"
new_data = np.genfromtxt(new_file,skip_header=2,usecols=(0,1))

#print(new_data)
#print(new_data.shape)
print("")

first_years = new_data[0:359,0]
second_years = new_data[359:718,0]

first_co2 = new_data[0:359,1]
second_co2 = new_data[359:718,1]

#First Graph

intercept_1, slope_1 = np.polynomial.polynomial.polyfit(first_years,first_co2,1)
print("Slope of CO2 levels from 1958 to 1988")
print(intercept_1,slope_1)
print("")

#create best fit line's x and y points
a = np.linspace(1958,1988)


#best fit line
b = slope_1*a + intercept_1

plt.xlabel("Year")
plt.ylabel("CO2 levels (ppm)")
plt.title("CO2 levels from 1958 to 1988")
plt.scatter(first_years,first_co2,color="purple")
plt.plot(a,b)

#Second Graph
intercept_2,slope_2 = np.polynomial.polynomial.polyfit(second_years,second_co2,1)


c = np.linspace(1988,2018)


#best fit line
d = slope_2*c + intercept_2

#CO2 estimated amount in the year 2040






```
